```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Preciousdisono/BSIS-2A-PRECIOUS-KEZIAH-DISONO/main/SALARY_DATA.csv')
```

```python
print(df)
```

```python
mean_salary = df['Salary'].mean()
```

```python
print(mean_salary)
```

```python
median_salary = df['Salary'].median()
```

```python
print(median_salary)
```

```python
std_deviationsalary = df['Salary'].std()
```

```python
print(std_deviationsalary)
```

```python
total_salary = df['Salary'].sum()
```

```python
print(total_salary)
```

```python
sns.histplot(df['Salary'])
```

```python
plt.show()
```

```python
sns.catplot(data=df,col = "Name",x = "Salary",kind = "count")
```

```python
df_p = df.groupby(['Name', 'Salary']).size().head(10)
```

```python
(df_p.head())
```

```python
top_5 = df_p.sort_values(ascending = True).head(5)
```

```python
df_p.plot(kind='bar')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```

```python
plt.show()
```

```python
top_5.plot(kind = 'pie')
plt.title('total salary')
plt.show()
```

```python
df_p.plot(kind = 'line')
plt.title('total salary')
plt.show()
```

```python
df_p = df.groupby([ 'Department']).size().head(10)
```

```python
df_p.plot(kind = 'line')
plt.title('Department')
plt.show()
```

```python
df_p.plot(kind = 'pie')
plt.title('Department')
plt.show()
```

```python
age_info = df[df['Age'].notnull()][['Age']]

#Dropping values of age = 0
age_info_clean = age_info[age_info['Age'] != 0]

age_info_clean.info()
```

```python
fig = plt.figure(figsize = (20,8))
Age_plot = sns.distplot(age_info_clean, bins = 30)
Age_plot.tick_params(labelsize = 13)
mean = age_info['Age'].mean()

plt.xlabel('Age', fontdict = {'fontsize':15})
plt.title('Age of disease-affected humans', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(mean,color = 'red')
plt.show()
```
