```python
sex_info = df[df['Sex'].notnull()][['Sex']]

#Dropping values of sex = 0
sex_info_clean = sex_info[sex_info['Sex'] != 0]
sex_info_clean.info()
```

```python
fig = plt.figure(figsize = (20,8))
Sex_plot = sns.distplot(sex_info_clean, bins = 30)
Sex_plot.tick_params(labelsize = 13)

std = sex_info['Sex'].std()
plt.xlabel('Sex', fontdict = {'fontsize':15})
plt.title('Sex of disease-affected humans', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(std,color = 'red')
plt.show()
```
